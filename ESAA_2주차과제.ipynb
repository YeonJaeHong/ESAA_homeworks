04-2. matplotlib 라이브러리 자유재로 사용하기


```python
#기초 그래프 그리기
#tips 데이터 집합을 불러와서 변수에 저장
tips= sns.load_dataset('tips')
print(tips.head())
print(type(tips))
```

```python
#기본 틀인 fig를 마련하고 그래프 격자를 구성
fig=plt.figure()
axes1=fig.add_subplot(1,1,1)
```

```python
#hist 메서드에 total_bill을 전달하면 히스토그램이 만들어짐, x축 간격은 bins 인잣값으로 조정할 수 있음
axes1.hist(tips['total_bill'],bins=10)
axes1.set_title('Histogram of Tatal Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
```

```python
#산점도 그리기. tip y변수, total bill: x변수
scatter_plot=plt.figure()
axes1= scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])
axes1.set_title('scatterplot of total bill vs tip')
axes1.set_xlabel('total_bill')
axes1.set_ylabel('Tip')
```

```python
#박스그래프
boxplot= plt.figure()
axes1= boxplot.add_subplot(1,1,1)
axes1.boxplot([tips[tips['sex']=='Female']['tip'],tips[tips['sex']=='Male']['tip']], labels=['Female','Male']
)
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by sex')

```

```python
#다변량 그래프 그리기
#성별을 새로운 변수로 추가하기
#여성은 0, 남성은 1
def recode_sex(sex):
  if sex=='Female':
    return 0
  else:
   return 1

```

```python
#recode_sex 가 반환한 값을 데이터프레임에 sex_color 열을 새로 생성 후 추가
tips['sex_color']= tips['sex'].apply(recode_sex)
```

```python
# 테이블당 인원수도 산점도에 추가하기
scatter_plot= plt.figure()
axes1= scatter_plot.add_subplot(1,1,1)
axes1.scatter (
    x=tips['total_bill'],
    y=tips['tip'],
    s= tips['size']*10,
    c=tips['sex_color'],
    alpha=0.5
)
axes1.set_title('total_bill vs tip colored by sex and by size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

04-3.seaborn 라이브러리 자유자재로 이용하기


```python
import seaborn as sns
tips= sns.load_dataset('tips')
```

```python
#seaborn 라이브러리로 히스토그램을 그리려면 subplots, displot 메서드 사용
ax=plt.subplots()
ax= sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
# kde= false,
# hist= false 로 인자를 지정하면 된다
```

다양한 종류의 이변량 그래프 그리기

```python
#1. seaborn 라이브러리로 산점도 그래프 그리기-> regplot method를 이용
ax= plt.subplots()
ax= sns.regplot(x='total_bill', y='tip', data=tips) #여기서 회귀선을 없애려면 fit_reg=False로 쓰면 된다
ax.set_title=('Scatterplot of total bill and tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')

```

```python
#2. jointplot 을쓰면 산점도와 히스토그램을 한번에 그려준다 x,y 인자에 원하는 열 이름을 지정하고 data 인잣값으로 데이터프레임지정
joint= sns.jointplot(x='total_bill',y='tip',data=tips) #추가적으로 kind='hex'를 쳐주면 육각형형태를 볼 수 있다
joint.set_axis_labels(xlabel='total_bill',ylabel='tip')
joint.fig.suptitle('Joint plot of total bill and tip', fontsize=10,y=1.03)
```

```python
#이차원 밀집도그리기 kdeplot method
ax=plt.subplots()
ax= sns.kdeplot(data=tips['total_bill'], data2= tips['tip'], fill=True) #??? 그래서 뭐가 문제인건데...
ax.set_title('Kernel desity plot of total bill and tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')

```

```python
#바그래프 그리기
ax= plt.subplots()
ax=sns.barplot(x='time', y='total_bill',data=tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

```python
#박스그래프
ax= plt.subplots()
ax=sns.boxplot(x='time', y='total_bill',data=tips)
ax.set_title('Boxplot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

```python
#violinplot
ax= plt.subplots()
ax=sns.violinplot(x='time', y='total_bill',data=tips)
ax.set_title('Violin plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('total bill')
```

```python
#관계 그래프 그리기; pariplot method
fig=sns.pairplot(tips)
#대각선기준으로 위쪽, 아래 등으로 그려짐
pair_grid= sns.PairGrid(tips)
pair_grid= pair_grid.map_upper(sns.regplot)
pair_grid= pair_grid.map_lower(sns.kdeplot)
pair_grid= pair_grid.map_diag(sns.distplot,rug=True)
plt.show()
```

```python
#다변량 그리기: 다양한 그래프에서 색상 추가하기
#1) violin
ax= plt.subplots()
ax=sns.violinplot(x='time', y='total_bill',hue='sex',data=tips,split=True) #hue 인잣값으로 색상에 사용할 열 이름 추가

```

```python
#2) 산점도, 관계 에서 색상 추가
ax=sns.lmplot(x='total_bill', y='tip',data=tips,hue='sex',fit_reg=False)
```

```python
fig=sns.pairplot(tips,hue='sex')
```

```python
#산점도 그래프의 크기와 모양 조절
#산점도 점 크기 조절: scatter_kws 에 딕셔너리 형태로 인잣값 전달
scatter = sns.lmplot(x='total_bill', y='tip',data=tips, fit_reg=False, hue='sex',scatter_kws={'s': tips['size']*10})
# 산점도 그래프의 점을 다른 기호로 표현하고 싶으면 markers 인자값에 넣어주면 된다
```

```python
#lmplot 메서도로 4개의 데이터 그룹에 대한 그래프 한번에 그리기
anscombe_plot =sns.lmplot(x='x', y='y',data=anscombe,fit_reg=False)
```

```python
#그룹별로 그래프를 나누어 그리려면 col_wrap ; 열의 최댓값, col=dataset열을 지정
anscombe_plot =sns.lmplot(x='x', y='y',data=anscombe,fit_reg=False,col='dataset', col_wrap=2)
```
