dataframe, series 정렬 : sort_values()
 >주요 입력 파라미터는 by, ascending, inplace

*   오름차순: ascending=True (기본)
*   내림차순: ascendign = False


*   inplace=False로 설정하면 sort_values()를 호출한 dataframe은 그대로 이고, 정렬된 dataframe을 결과로 반환한다 (기본)
*  inplace=True 로 설정하면 호출한 dataframe의 정렬 결과를 그대로 적용





```python
#이름으로 오름차순
titanic_sorted = titanic_df.sort_values(by=['Name'])
titanic_sorted.head(3)
```

```python
#여러개의 칼럼으로 정렬하려면 by에 리스트 형식으로 정렬하려는 칼럼을 입력하면 된다
titanic_sorted= titanic_df.sort_values(by=['Pclass','Name'],ascending=False)
titanic_sorted.head(3)
```

결손 데이터 처리하기
> 결손 데이터는 칼럼에 값이 없는 null인 경우를 말한다.


*  numpy 의 NaN으로 표시.머신러닝 알고리즘은 이 값을 처리하지 않으므로 다른 값으로 대체해야한다
*  이 nan 값은 평균, 총합등의 함수 연산시 제외 된다.


*   NaN의 여부를 확인하는 API는 isna()
*   NaN값을 다른 값으로 대체하는 API는 fillna()이다

```python
# isna()로 결손 데이터 확인
titanic_df.isna().head(3)
```

```python
#결손 데이터의 개수를 구하려면 sum을 추가하면된다.
titanic_df.isna().sum()
```

```python
#fillna()로 결손 데이터 대체하기
#타이타닉 데이터 세트의 'cabin'칼럼의 NaN값을 'C000'으로 대체
titanic_df['Cabin']= titanic_df['Cabin'].fillna('C000')
titanic_df.head(3)
```

```python
#주의해야할점 : fillna()를 이용해서 반환값을 다시 받거나 inplace=True 파라미터를 fillna()에 추가해야 실제 데이터세트값이 변경된다
titanic_df['Cabin']= titanic_df['Cabin'].fillna('C000',inplace=True) #이렇게 해야 실제 데이터세트의 값이 변경된다
```

```python
#'Age' 칼럼의 NaN 값을 평균나이로, 'Embarked' 칼럼의 Nan 값을 'S'로 대체해 모든 결손 데이터를 처리한다
titanic_df['Age']=titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df['Embarked']=titanic_df['Embarked'].fillna('S')
titanic_df.isna().sum()
```
